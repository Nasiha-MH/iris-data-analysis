# Visualize the dataset

Create a scatter plot to show relationships between features.

```python
import matplotlib.pyplot as plt

# Scatter plot: relationship between two features
sns.scatterplot(x='SepalLengthCm', y='PetalLengthCm', hue='Species', data=df)
plt.title("Sepal Length vs Petal Length")
plt.show()
```

Use histograms to show value distributions.

```python
# Histograms for all numeric features
df.hist(figsize=(10, 8))

plt.suptitle("Distribution of Features")
plt.show()
```

Use box plots to identify outliers.

```python
# Box plot for all numeric features
sns.boxplot(data=df)

plt.title("Box Plot for Outlier Detection")
plt.show()
```
